```python
from ultralytics import YOLO
```

```python
model = YOLO(r"D:\Final_Thesis\yolov12mepoch30\runs_backup2\detect\yolov12m_brsdd\weights\best.pt")
```

```python
results = model.val(
    data=r"D:\Final_Thesis\yolov12mepoch30\BRSDD\data.yaml",
    split='test'      # <– IMPORTANT
)
```

```python
precision = results.box.p.mean()
recall = results.box.r.mean()
map50 = results.box.ap50.mean()
map5095 = results.box.ap.mean()

print(f"Precision:  {precision:.4f}")
print(f"Recall:     {recall:.4f}")
print(f"mAP50:      {map50:.4f}")
print(f"mAP50-95:   {map5095:.4f}")
```
